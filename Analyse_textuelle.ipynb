```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Supposons que df_AI_subject est déjà défini et contient la colonne 'Description/links'
urls = df_AI_subject['Description/links'].tolist()

# Créer une liste pour stocker les textes extraits
textes_extraits = []

# Fonction pour extraire et nettoyer le texte d'un lien
def extract_and_clean_text(url):
    try:
        # Faire une requête HTTP pour récupérer le contenu de la page
        response = requests.get(url, timeout=10)  # Ajouter un timeout pour éviter les longues attentes
        response.raise_for_status()  # Vérifier si la requête a réussi

        # Analyser le contenu HTML avec BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extraire tout le texte de la page
        all_text = soup.get_text()

        # Séparer le texte en lignes et supprimer les espaces supplémentaires
        clean_text = '\n'.join(line.strip() for line in all_text.splitlines() if line.strip())

        # Retourner le texte nettoyé
        return clean_text
    except requests.exceptions.RequestException as e:
        print(f"Erreur lors de la requête pour l'URL {url}: {e}")
        return None

# Appliquer la fonction à chaque lien
for url in urls:
    try:
        html = requests.get(url, timeout=10)
        html.raise_for_status()
        soup = BeautifulSoup(html.text, 'html.parser')
        links_wiki = soup.find_all('ul', class_="n8H08c")
        mylist = []
        for ul in links_wiki:
            links = ul.find_all('a', href=True)
            for link in links:
                mylist.append(link['href'])

        # Créer une liste pour stocker les textes extraits de chaque page
        textes_page = []

        # Appliquer la fonction extract_and_clean_text à chaque lien
        for sub_url in mylist:
            if sub_url.startswith('https://'):
                texte = extract_and_clean_text(sub_url)
                if texte:
                    textes_page.append(texte)
                    print(f"Texte extrait de: {sub_url}")
                    print("----------------------------------------")
                else:
                    print(f"Échec de l'extraction de texte pour: {sub_url}")
            else:
                print(f"URL non valide: {sub_url}")
    except requests.exceptions.RequestException as e:
        print(f"Erreur lors de la requête pour l'URL principale {url}: {e}")
        continue

    # Concaténer les textes extraits de la page en une seule chaîne
    texte_concatene = '\n\n'.join(textes_page)
    
    # Ajouter le texte concaténé à la liste des textes extraits
    textes_extraits.append(texte_concatene)

# Exemple d'utilisation des textes extraits (par exemple, les afficher)
for i, texte in enumerate(textes_extraits):
    print(f"Texte extrait {i+1}:")
    print(texte)
    print("="*80)
```
